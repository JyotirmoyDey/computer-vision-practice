```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('hand.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
image_gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
retval, binary_image = cv2.threshold(image_gray, 225, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary_image, cmap='gray')
```

<h3>Draw contours</h3>

```python
retval, contours, hierarchy = cv2.findContours(binary_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

image_display = np.copy(image_copy)
all_contours = cv2.drawContours(image_display, contours, -1, (0, 255, 0), 2)
plt.imshow(all_contours)
```
