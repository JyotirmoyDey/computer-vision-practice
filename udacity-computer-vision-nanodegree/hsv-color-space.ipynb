```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('water_balloons.jpg')

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
# RGB channels
red_channel = image_copy[:, :, 0]
green_channel = image_copy[:, :, 1]
blue_channel = image_copy[:, :, 2]
```

```python
# lets plot each channel in gray scale
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20, 10))
ax1.set_title('red')
ax1.imshow(red_channel, cmap='gray')

ax2.set_title('green')
ax2.imshow(green_channel, cmap='gray')

ax3.set_title('blue')
ax3.imshow(blue_channel, cmap='gray')
```

```python
# convert the image into hsv color space
hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)
h = hsv[:, :, 0]
s = hsv[:, :, 1]
v = hsv[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20, 10))
ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('value')
ax3.imshow(v, cmap='gray')
```

The hue has high values for pink color

<h3> Define thresholds for pink color and hue channel </h3>

```python
pink_low = np.array([180, 0, 100])
pink_high = np.array([255, 255, 230])

hue_low = np.array([160, 0,0])
hue_high = np.array([180, 255, 255])
```

<h3>Mask the image</h3>

```python
# RGB mask
mask_rgb = cv2.inRange(image_copy, pink_low, pink_high)
masked_pink_image = np.copy(image_copy)
masked_pink_image[mask_rgb == 0] = [0,0,0]

plt.imshow(masked_pink_image)
```

```python
# HSV mask
mask_hsv = cv2.inRange(hsv, hue_low, hue_high)
masked_hsv_image = np.copy(image_copy)
masked_hsv_image[mask_hsv == 0] = [0,0,0]

plt.imshow(masked_hsv_image)
```

</h4>Thus we can see that an HSV space is more valuable in selecting an area nder varying light conditions<h4>
