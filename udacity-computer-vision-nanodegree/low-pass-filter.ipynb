<h3>Gaussian blur filter</h3>
blurs both the image and also detects the edges

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('brain_MR.jpg')
plt.imshow(image)

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
# convert to grayscale
image_gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# create a gaussian blurred image
gray_blur = cv2.GaussianBlur(image_gray, (5,5), 0)

plt.imshow(gray_blur, cmap='gray')
```
