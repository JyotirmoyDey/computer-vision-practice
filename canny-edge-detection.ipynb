<h3>Canny edge detection</h3>
Widely used and accurate edge detection algorithm
<ul>
    <li>Filters out noise using a Gaussian blur</li>
    <li>Finds the strength and direction of edges using Sobel filters</li>
    <li>Applies non-maximum suppression to isolate the strongest edges and thin them to one-pixel wide lines</li>
    <li>Uses hysteresisto isolate the best edges</li>
</ul>


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('sunflower.jpg')

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
image_gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap='gray')
```

<h3>Implement canny edge detection</h3>

```python
# define lower and upper thresholds for hysteresis
lower = 120
upper = 240

edges = cv2.Canny(image_gray, lower, upper)
plt.imshow(edges, cmap='gray')
```
