```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('pizza_bluescreen.jpg')
print ("Image shape, ", image.shape)
plt.imshow(image)
```

```python
# make a copy of the image
image_copy = np.copy(image)
# change color from BGR to RGB
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
# Display
plt.imshow(image_copy)
```

<h3>Define a colour threshold</h3>

```python
# define the lower and upper threshold for selecting the blue area
lower = np.array([0,0,220])
upper = np.array([60,70, 255])
```

<h3>Create a mask</h3>

```python
# define the masked range
mask = cv2.inRange(image_copy, lower, upper)
plt.imshow(mask, cmap='gray')
print(mask.shape)
```

```python
masked_image = np.copy(image_copy)
masked_image[mask !=0] = [0,0,0]
# Display 
plt.imshow(masked_image)
```

```python
# apply a background to this image
background_image = cv2.imread('space_background.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)
cropped_background = background_image[0:514, 0:816]

cropped_background[mask == 0] = [0,0,0]
plt.imshow(cropped_background)
```

```python
complete_image = cropped_background + masked_image
plt.imshow(complete_image)
```
