```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('triangle_tile.jpeg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
image_gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(image_gray, cmap='gray')
```

<h3>Detect Corners</h3>

```python
image_gray = np.float32(image_gray)

corners = cv2.cornerHarris(image_gray, 2, 3, 0.04)
corners = cv2.dilate(corners, None)


plt.imshow(corners, cmap='gray')
```

<h3>Select and display strongest corners</h3>

```python
# Define a threshold for extracting strong corners
threshold = 0.01 * corners.max()

corner_image = np.copy(image_copy)
for j in range(0, corners.shape[0]):
    for i in range(0, corners.shape[1]):
        if(corners[j,i] > threshold):
            cv2.circle(corner_image, (i, j), 2, (0, 255, 0), 1)
            
plt.imshow(corner_image)
```
